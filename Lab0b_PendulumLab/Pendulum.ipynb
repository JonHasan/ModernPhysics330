# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt 
```

# Make Data 

```python
x_data = np.array([np.sqrt(16), np.sqrt(25), np.sqrt(36)]) 

y_data = np.array([5.534 / 6,6.628 / 6, 7.87/6 ])

```

```python
intercept = 0 

slope = (2* np.pi)/np.sqrt(980.665)

slope
```

# Plot the data 

Choose appropriate uncertanties, x should have uncertainty of .5 and y should have uncertainty of .005 

```python
plt.figure()

fig, ax = plt.subplots()

plt.scatter(x_data, y_data)

ax.axline((0, intercept), slope = slope, color = "red" )

plt.errorbar(x_data, y_data, xerr =.5 ,yerr = .005, ls = 'none')

plt.xlabel("Square root of string length (cm)")

plt.ylabel("Period (s)")

plt.title("Period vs square root of string length")

plt.savefig("PendulumData.png")
```

## Correction 

It came to my  attention that I forgot a key piece of information. The length in the equation is from the top of the pendulum to the center of mass of the weight. I need to find the distance of the midpoint of the weight and add it to the string length. 

```python
dist_COM = 1.4 

x_data = np.array([np.sqrt(16 + dist_COM), np.sqrt(25 + dist_COM), np.sqrt(36 + dist_COM)]) 
```

# Plot Corrected Data 

```python
plt.figure()

fig, ax = plt.subplots()

plt.scatter(x_data, y_data)

ax.axline((0, intercept), slope = slope, color = "red" )

plt.errorbar(x_data, y_data, xerr =.5 ,yerr = .001, ls = 'none')

plt.xlabel("Square root of Length to weight COM (cm)")

plt.ylabel("Period (s)")

plt.title("Period vs square root of length to weight COM")

plt.savefig("PendulumData_Corrected.png")
```

With the error bars, the trend line values are included within the error bars. This means that this data is good to use.
